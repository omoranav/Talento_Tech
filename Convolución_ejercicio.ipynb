```python
import numpy as np
import matplotlib.pyplot as plt

def plot_convolution(X, h, X_zero_index, h_zero_index):
    # Calcular la convolución de X y h
    y = np.convolve(X, h)

    # Definir los índices de las secuencias para la gráfica
    X_indices = np.arange(-X_zero_index, len(X) - X_zero_index)
    h_indices = np.arange(-h_zero_index, len(h) - h_zero_index)
    y_indices = np.arange(X_indices[0] + h_indices[0], X_indices[-1] + h_indices[-1] + 1)

    # Graficar las secuencias
    plt.figure(figsize=(8, 6))

    plt.subplot(3, 1, 1)
    plt.stem(X_indices, X, linefmt='b-', markerfmt='bo', basefmt='r-')
    plt.plot(X_indices, X, 'b', linestyle='-', marker='o')  # Unir los puntos con una línea
    plt.title('Secuencia X')
    plt.xlabel('n')
    plt.ylabel('X[n]')
    plt.grid()

    plt.subplot(3, 1, 2)
    plt.stem(h_indices, h, linefmt='b-', markerfmt='bo', basefmt='r-')
    plt.plot(h_indices, h, 'b', linestyle='-', marker='o')  # Unir los puntos con una línea
    plt.title('Secuencia h')
    plt.xlabel('n')
    plt.ylabel('h[n]')
    plt.grid()

    plt.subplot(3, 1, 3)
    plt.stem(y_indices, y, linefmt='b-', markerfmt='bo', basefmt='r-')
    plt.plot(y_indices, y, 'b', linestyle='-', marker='o')  # Unir los puntos con una línea
    plt.title('Convolución de X y h')
    plt.xlabel('n')
    plt.ylabel('y[n]')
    plt.grid()

    plt.tight_layout()
    plt.show()

# Definir las secuencias
X = np.array([1, 2, 3, 4, 5])
h = np.array([-1, 5, 3, -2, 1])

# Definir los índices donde el valor cero está ubicado
X_zero_index = 1
h_zero_index = 2

# Llamar a la función para graficar la convolución
plot_convolution(X, h, X_zero_index, h_zero_index)
```

```python
#Código para hacer el cálculo de convolución de manera manual

import numpy as np
import matplotlib.pyplot as plt

def manual_convolution(X, h):
    len_X = len(X)                 # Longitud de la secuencia X
    len_h = len(h)                 # Longitud de la secuencia h
    y_length = len_X + len_h - 1   #Calcular la longitud de la secuencia resultante
    y = np.zeros(y_length)         # Crear un arreglo de ceros para la convolución

    # Realizar la convolución manualmente

    # Bucle for que recorre todos los elementos de las secuencias X y h
      # En cada iteración del bucle interno, se realiza la multiplicación del
      # elemento X[i] por el elemento h[j] y se suma al elemento correspondiente
      # de y en la posición i + j.

    for i in range(len_X):
        for j in range(len_h):
            y[i + j] += X[i] * h[j]

    return y

def plot_convolution(X, h, X_zero_index, h_zero_index):
    # Calcular la convolución manual de X y h
    y = manual_convolution(X, h)

    # Definir los índices de las secuencias para la gráfica
    X_indices = np.arange(-X_zero_index, len(X) - X_zero_index)
    h_indices = np.arange(-h_zero_index, len(h) - h_zero_index)
    y_indices = np.arange(X_indices[0] + h_indices[0], X_indices[-1] + h_indices[-1] + 1)

    # Graficar las secuencias
    plt.figure(figsize=(8, 6))

    plt.subplot(3, 1, 1)
    plt.stem(X_indices, X, linefmt='b-', markerfmt='bo', basefmt='r-')
    plt.plot(X_indices, X, 'b', linestyle='-', marker='o')  # Unir los puntos con una línea
    plt.title('Secuencia X')
    plt.xlabel('n')
    plt.ylabel('X[n]')
    plt.grid()

    plt.subplot(3, 1, 2)
    plt.stem(h_indices, h, linefmt='b-', markerfmt='bo', basefmt='r-')
    plt.plot(h_indices, h, 'b', linestyle='-', marker='o')  # Unir los puntos con una línea
    plt.title('Secuencia h')
    plt.xlabel('n')
    plt.ylabel('h[n]')
    plt.grid()

    plt.subplot(3, 1, 3)
    plt.stem(y_indices, y, linefmt='b-', markerfmt='bo', basefmt='r-')
    plt.plot(y_indices, y, 'b', linestyle='-', marker='o')  # Unir los puntos con una línea
    plt.title('Convolución de X y h')
    plt.xlabel('n')
    plt.ylabel('y[n]')
    plt.grid()

    plt.tight_layout()
    plt.show()

# Definir las secuencias
X = np.array([1, 2, 3, 4, 5])
h = np.array([-1, 5, 3, -2, 1])

# Definir los índices donde el valor cero está ubicado
X_zero_index = 1
h_zero_index = 2

# Llamar a la función para graficar la convolución
plot_convolution(X, h, X_zero_index, h_zero_index)
```
