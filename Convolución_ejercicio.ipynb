```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las secuencias
X = np.array([1, 2, 3, 4, 5])
h = np.array([-1, 5, 3, -2, 1])

# Definir los índices donde el valor cero está ubicado
X_zero_index = 1
h_zero_index = 2

# Calcular la convolución de X y h
y = np.convolve(X, h)

# Definir los índices de las secuencias para la gráfica
X_indices = np.arange(-X_zero_index, len(X) - X_zero_index)
h_indices = np.arange(-h_zero_index, len(h) - h_zero_index)
y_indices = np.arange(X_indices[0] + h_indices[0], X_indices[-1] + h_indices[-1] + 1)

# Graficar las secuencias
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.stem(X_indices, X, use_line_collection=True)
plt.title('Secuencia X')
plt.xlabel('n')
plt.ylabel('X[n]')
plt.grid()

plt.subplot(3, 1, 2)
plt.stem(h_indices, h, use_line_collection=True)
plt.title('Secuencia h')
plt.xlabel('n')
plt.ylabel('h[n]')
plt.grid()

plt.subplot(3, 1, 3)
plt.stem(y_indices, y, use_line_collection=True)
plt.title('Convolución de X y h')
plt.xlabel('n')
plt.ylabel('y[n]')
plt.grid()

plt.tight_layout()
plt.show()
```
