c) fetch_openml:
Descripción: Descarga y carga conjuntos de datos desde OpenML.


```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784')
X, y = mnist.data, mnist.target
print(X)
print(y)
```
