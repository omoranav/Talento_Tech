```python
#Ejemplo de c贸mo realizar la convoluci贸n de una imagen
#utilizando OpenCv en Python

import numpy as np
import cv2
from scipy.signal import convolve
from google.colab.patches import cv2_imshow

#Cargar la imagen
image = cv2.imread('image.jpg')

# Redimensionar la imagen
new_width = 200
new_height = 300
image_resized = cv2.resize(image, (new_width, new_height))

#Definir el kernel de la convoluci贸n
kernel = np.array([[0, 1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

#Realizar la convoluci贸n
Colvolved_image = cv2.filter2D(image_resized, -1, kernel)

#Mostrar la imagen original y la imagen convolucionada
print("Original Image")
cv2_imshow(image_resized)
print("Colvolved Image")
cv2_imshow(Colvolved_image)
```
