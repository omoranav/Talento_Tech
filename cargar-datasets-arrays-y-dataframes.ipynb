## Conjuntos de Datos Preexistentes:

Para una base de datos en una url particular y manipulando datos con dataframes y archivos csv


- Iris dataset:

```python
import requests
import pandas as pd
import io

# URL of the Iris dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

# Perform GET request to fetch the data
response = requests.get(url)

# Define column names for the Iris dataset
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Read the dataset into a DataFrame
df = pd.read_csv(io.StringIO(response.text), header=None, names=column_names)

# Save the DataFrame to a CSV file
df.to_csv("iris_data.csv", index=False)

# Verify by reading the CSV file back into a DataFrame and printing the first few rows
df_check = pd.read_csv("iris_data.csv")
#print(df_check.head())

print(df.describe())
print(df_check.describe())
print(df)
df_check
```

## As a dataframe from arrays:

- Manuscript digits dataset (itÂ´s NOT a large dataset):

This is a copy of the test set of the UCI ML hand-written digits datasets:

https://archive.ics.uci.edu/ml/datasets/Optical+Recognition+of+Handwritten+Digits

(See description below)

```python
import pandas as pd
from sklearn.datasets import load_digits

# Load the digits dataset
digits = load_digits()

# Get the features (X) and target labels (y)
X, y = digits.data, digits.target

# Create a DataFrame from the feature data
digits_df = pd.DataFrame(X)

# Add the target labels as a new column in the DataFrame
digits_df['label'] = y

# Display the first few rows of the DataFrame
#print(digits_df.head())

#print(digits_df.describe())

# InformaciÃ³n adicional sobre el dataset
print(f"Dimensiones de X: {X.shape}")
print(f"Dimensiones de y: {y.shape}")
print(f"Nombres de las caracterÃ­sticas: {digits.feature_names}")
print(f"DescripciÃ³n del dataset: {digits.DESCR}")

digits_df
```

# MNIST: large manuscript digits dataset
(The MNIST database (Modified National Institute of Standards and Technology database) is a large database of handwritten digits that is commonly used for training various image processing systems. The database is also widely used for training and testing in the field of machine learning.)

Usando fetch_openml: descarga conjuntos de datos desde OpenML.

```python
import pandas as pd
from sklearn.datasets import fetch_openml

# Fetch the MNIST dataset
mnist = fetch_openml('mnist_784', version=1)

# Get the features (X) and target labels (y)
X, y = mnist.data, mnist.target

# Create a DataFrame from the feature data
mnist_df = pd.DataFrame(X)

# Add the target labels as a new column in the DataFrame
mnist_df['label'] = y

# Display the first few rows of the DataFrame
#print(mnist_df.head())

#print(mnist_df.describe())
mnist_df
```

4. APIs (Interfaces de ProgramaciÃ³n de Aplicaciones):
a) Acceso a una API PÃºblica:


```python
import requests

# URL de la API pÃºblica
url = "https://jsonplaceholder.typicode.com/posts/1"

# Realizar la solicitud GET a la API
response = requests.get(url)

# Verificar si la solicitud fue exitosa (cÃ³digo de estado 200)
if response.status_code == 200:
    # Convertir la respuesta a formato JSON
    data = response.json()

    # Extraer la informaciÃ³n relevante
    relevant_info = data["title"]

    # Imprimir la informaciÃ³n relevante
    print(relevant_info)
else:
    # Imprimir un mensaje de error si la solicitud no fue exitosa
    print("Error al realizar la solicitud a la API")

data
```
